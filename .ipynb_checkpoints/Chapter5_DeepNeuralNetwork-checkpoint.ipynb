```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## Easy Data

```python
x1 = np.arange(0, 21)
y1 = 2*x1 + 1
```

```python
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

```python
model1.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss='mse',
    metrics=['mse']
)
```

```python
history1 = model1.fit(x1, y1, epochs=1000)
```

```python
y_pred1 = model1.predict(x1)
```

```python
plt.title('Easy Data trained with a single neuron')
plt.scatter(x1, y1)
plt.plot(y_pred1, color='red')
plt.show()
```

## Slightly Hard Data

```python
x2 = np.arange(0, 21)
y2 = 2*x2**2 + 1
```

```python
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
model2.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mse',
    metrics=['mse']
)
```

```python
history2 = model2.fit(x2, y2, epochs=1000)
```

```python
y_pred2 = model2.predict(x2)
```

```python
plt.title('Slightly Hard Data trained with a single neuron')
plt.scatter(x2, y2)
plt.plot(y_pred2, color='red')
plt.show()
```

## Using Deep Neural network

```python
dnn = tf.keras.Sequential([
    tf.keras.layers.Dense(5, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1)
])

dnn.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mse',
    metrics=['mse']
)
```

```python
history = dnn.fit(x2, y2, epochs=2000)
```

```python
y_pred_dnn = dnn.predict(x2)
```

```python
plt.title('Slightly Hard Data trained with a DNN')
plt.scatter(x2, y2)
plt.plot(y_pred_dnn, color='red')
plt.show()
```

### Using more complex DNN

```python
dnn_complex = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1)
])

dnn_complex.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mse']
)
```

```python
history = dnn_complex.fit(x2, y2, epochs=2000)
```

```python
y_pred_dnn_complex = dnn_complex.predict(x2)
```

```python
plt.title('Slightly Hard Data trained with a more complex DNN')
plt.scatter(x2, y2)
plt.plot(y_pred_dnn_complex, color='red')
plt.show()
```
